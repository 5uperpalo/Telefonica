```python
import pandas as pd
import seaborn as sns
import numpy as np
import sklearn
sns.set_style("whitegrid")
import pymysql.cursors
import datetime
import matplotlib
#matplotlib.use('Agg')
%matplotlib nbagg
import matplotlib.pyplot as plt
from pylab import *
import pickle
import itertools
```

```python
datafeb = pd.read_csv("201902", sep=',',header=None)
datamarch=pd.read_csv("201903", sep=',',header=None)
lsoemapping=pd.read_csv("../lsoe-mapping/LondonAntennaGenerationIMD_v2.csv", sep=',')
```

```python
datafeb=datafeb.rename(columns = {1: "lkey"})
datamarch=datamarch.rename(columns = {1: "lkey"})
```

```python
datafeb=pd.merge(datafeb,lsoemapping,on='lkey')
datamarch=pd.merge(datamarch,lsoemapping,on='lkey')
data=pd.concat([datafeb,datamarch])
```

```python
data_avg=data.groupby(["lkey","LSOA_IMD_decile","gen"]).mean()
```

```python
font = {'family' : 'Helvetica',
        'weight' : 'medium',
        'size'   : 16}
matplotlib.rc('font', **font)
plt.figure()
sns.boxplot(x="LSOA_IMD_decile", y=43, hue="gen", data=data_avg.reset_index(), showfliers=False, whis=[10,90], palette="Set1")
plt.legend(loc='upper left')
plt.ylabel("Data Exchanged [#]")
plt.xlabel("IMD decile")
plt.tight_layout()
savefig('byte-density-mix-tecno.pdf')

plt.show()
```

```python
font = {'weight' : 'medium',
        'size'   : 25}
matplotlib.rc('font', **font)
max_value=max(data_avg.groupby("LSOA_IMD_decile")[43].quantile(0.9))
data_avg_normalized=data_avg/max_value
fig, ax = plt.subplots(figsize=(10,5))

data_avg_normalized.reset_index().boxplot(column=43,by="LSOA_IMD_decile",showfliers=False, whis=[10,90], ax=ax)
plt.title("")
plt.suptitle("")
ax.set_ylabel("Data Exchanged")
ax.set_xlabel("IMD decile")
plt.tight_layout()
#savefig('byte-exchanged-sector.pdf')
```

```python
font = {'weight' : 'medium',
        'size'   : 25}
matplotlib.rc('font', **font)
max_value=max(data_avg.groupby("LSOA_IMD_decile")[43].quantile(0.9))
data_avg_normalized=data_avg/max_value
fig, ax = plt.subplots(figsize=(10,5))

data_to_plot=data_avg_normalized.reset_index()[data_avg_normalized.reset_index().gen=="4G"]
data_to_plot.boxplot(column=43,by="LSOA_IMD_decile",showfliers=False, whis=[10,90], ax=ax)
plt.title("")
plt.suptitle("")
ax.set_ylabel("Data Exchanged")
ax.set_xlabel("IMD decile")
plt.tight_layout()
savefig('byte-exchanged-sector-4G.pdf')
```

```python
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

data_avg_anova=data_avg.reset_index().rename(columns={43: "tot_bytes"})

anova_results = ols('tot_bytes ~ C(LSOA_IMD_decile)', data=data_avg_anova).fit()
print anova_results.summary()



aov_table = sm.stats.anova_lm(anova_results, typ=2)
print aov_table



from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison

mc = MultiComparison(data_avg_anova['tot_bytes'], data_avg_anova['LSOA_IMD_decile'])
mc_results = mc.tukeyhsd()
print(mc_results)
```
